```python
# Complete integrated script: data loading, algorithms, comparison plots, and normalized spider chart with execution time

from UHI_Function import *
import time
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# --- 1. Load data and initial stats ---
type_matrix, temp_matrix, type_pivot, type_to_num, num_to_type = load_map('sample_grid.csv')
H, W = type_matrix.shape

# Assumes budget_max and tree_species are defined in UHI_Function or earlier in script
# e.g., budget_max = 1000
#       tree_species = {'oak': {'base_price': 10, 'color':'green'}, ...}

vmin, vmax = temp_matrix.min(), temp_matrix.max()
baseline_avg, baseline_hot = calculate_objective_stats(temp_matrix, temp_matrix)
print(f"Baseline avg temperature: {baseline_avg:.2f}")
print(f"Baseline hotspot temperature: {baseline_hot:.2f}")

# Initial heat island map
generate_heat_island_map(temp_matrix, temp_matrix, threshold=32, show=True)

# --- 2. Random Insertion heuristic ---
def random_insertion(type_matrix, budget_max, type_to_num, tree_species, cost_factor=1.5):
    H, W = type_matrix.shape
    placed, total_cost = [], 0
    min_cost = min(tree_species[t]['base_price'] for t in tree_species)
    used = set()
    max_iters = H * W * 10
    iters = 0

    while total_cost + min_cost <= budget_max and iters < max_iters:
        iters += 1
        r, c = random.randrange(H), random.randrange(W)
        if type_matrix[r, c] == type_to_num['building'] or (r, c) in used:
            continue
        species = random.choice(list(tree_species.keys()))
        factor = cost_factor if type_matrix[r, c]==type_to_num['road'] else 1.0
        cost = tree_species[species]['base_price'] * factor
        if total_cost + cost > budget_max:
            continue
        placed.append((r, c, species))
        used.add((r, c))
        total_cost += cost

    return placed

# --- 3. Genetic Algorithm wrapper ---
def run_ga(type_matrix, temp_matrix, budget_max, type_to_num, tree_species,
           pop_size=300, max_iter=150, mutation_prob=0.1):
    dim = 4 * 50
    H, W = type_matrix.shape
    varbound = []
    for _ in range(50):
        varbound.extend([
            (0, H-1),         # row
            (0, W-1),         # col
            (0, len(tree_species)-1),  # species
            (0, 1)            # active flag
        ])
    varbound = np.array(varbound)

    def ga_obj(X):
        sol = []
        for i in range(0, len(X), 4):
            if int(X[i+3]) == 1:
                sol.append((int(X[i]), int(X[i+1]), int(X[i+2])))
        return objective_function(sol, type_matrix, temp_matrix, tree_species, type_to_num, budget_max)

    model = ga(
        function=ga_obj,
        dimension=dim,
        variable_type='int',
        variable_boundaries=varbound,
        algorithm_parameters={
            'max_num_iteration': max_iter,
            'population_size': pop_size,
            'mutation_probability': mutation_prob,
            'elit_ratio': 0.1,
            'crossover_probability': 0.7,
            'parents_portion': 0.3,
            'crossover_type': 'uniform',
            'max_iteration_without_improv': 10
        }
    )
    model.run()
    return model

# --- 4. Ant Colony Optimization wrapper ---
def update_pos_pheromone(best_sol, pher, evap=0.2, reinforcement=0.3):
    pher *= (1-evap)
    for (r, c, _) in best_sol:
        pher[r, c] += reinforcement
    return pher

def update_cand_pheromone(best_sol, cand_ph, evap=0.05, reinforcement=0.01):
    cand_ph *= (1-evap)
    for (_, _, sp) in best_sol:
        cand_ph[sp] += reinforcement
    return cand_ph

def probabilistic_selection(type_matrix, pher, cand_ph, budget_rem, cost_factor, type_to_num):
    choices = []
    H, W = type_matrix.shape
    for r in range(H):
        for c in range(W):
            if type_matrix[r, c] == type_to_num['building']:
                continue
            factor = cost_factor if type_matrix[r, c]==type_to_num['road'] else 1.0
            for sp in range(len(tree_species)):
                cost = tree_species[sp]['base_price'] * factor
                if cost <= budget_rem:
                    choices.append((sp, (r, c), pher[r, c] * cand_ph[sp]))
    if not choices:
        return None, None
    spp, pos, probs = zip(*choices)
    idx = random.choices(range(len(choices)), weights=probs, k=1)[0]
    return spp[idx], pos[idx]

def run_aco(type_matrix, temp_matrix, type_to_num, budget_max,
            iterations=100, num_ants=30, cost_factor=1.5,
            evap_pos=0.2, reinf_pos=0.3, evap_cand=0.05, reinf_cand=0.01,
            use_cand_ph=False):
    H, W = type_matrix.shape
    pher = np.ones((H, W))
    cand_ph = np.ones(len(tree_species))
    min_cost = min(tree_species[t]['base_price'] for t in tree_species)
    best_sol, best_q = [], np.inf

    for _ in range(iterations):
        for _ in range(num_ants):
            sol, rem = [], budget_max
            used = set()
            while rem >= min_cost:
                sp, pos = probabilistic_selection(type_matrix, pher, cand_ph, rem, cost_factor, type_to_num)
                if sp is None or pos in used: break
                r, c = pos
                cost = tree_species[sp]['base_price'] * (cost_factor if type_matrix[r, c]==type_to_num['road'] else 1.0)
                if rem < cost: break
                sol.append((r, c, sp))
                used.add(pos)
                rem -= cost
            cooled = apply_cooling(sol, temp_matrix, tree_species)
            q = np.mean(cooled) + 0.5 * np.max(cooled)
            if q < best_q:
                best_q, best_sol = q, sol
        pher = update_pos_pheromone(best_sol, pher, evap_pos, reinf_pos)
        if use_cand_ph:
            cand_ph = update_cand_pheromone(best_sol, cand_ph, evap_cand, reinf_cand)
    return best_sol

# --- 5. Single-run comparisons (heatmaps) ---
random_sol = random_insertion(type_matrix, budget_max, type_to_num, tree_species)
ga_model = run_ga(type_matrix, temp_matrix, budget_max, type_to_num, tree_species)
ga_sol = [(int(ga_model.output_dict['variable'][i]),
           int(ga_model.output_dict['variable'][i+1]),
           int(ga_model.output_dict['variable'][i+2]))
          for i in range(0, len(ga_model.output_dict['variable']), 4)
          if int(ga_model.output_dict['variable'][i+3])==1]
aco_sol = run_aco(type_matrix, temp_matrix, type_to_num, budget_max, iterations=100, num_ants=10)

# Compare heatmaps: initial, random, GA, ACO
fig, axs = plt.subplots(1, 4, figsize=(24,5))
# Initial
generate_heatmap(temp_matrix, vmin=vmin, vmax=vmax, show=False, ax=axs[0])
axs[0].set_title("Initial")
# Random
rand_temp = calculate_reduced_heatmap(random_sol, temp_matrix, tree_species)
generate_heatmap(rand_temp, vmin=vmin, vmax=vmax, show=False, ax=axs[1])
axs[1].set_title("Random")
# GA
ga_temp = calculate_reduced_heatmap(ga_sol, temp_matrix, tree_species)
generate_heatmap(ga_temp, vmin=vmin, vmax=vmax, show=False, ax=axs[2])
axs[2].set_title("GA")
# ACO
aco_temp = calculate_reduced_heatmap(aco_sol, temp_matrix, tree_species)
generate_heatmap(aco_temp, vmin=vmin, vmax=vmax, show=False, ax=axs[3])
axs[3].set_title("ACO")
plt.tight_layout()
plt.show()

# --- 6. Multi-run evaluation with execution time ---
def evaluate_multiple_runs(runs, type_matrix, temp_matrix, budget_max, type_to_num, tree_species):
    algs = ['Random', 'GA', 'ACO', 'ACO + Cand Ph']
    stats = {alg: [] for alg in algs}
    base_avg, base_hot = calculate_objective_stats(temp_matrix, temp_matrix)
    for _ in range(runs):
        # Random
        t0 = time.perf_counter()
        sol = random_insertion(type_matrix, budget_max, type_to_num, tree_species)
        t1 = time.perf_counter()
        used = calculate_used_budget(sol, type_matrix, tree_species, type_to_num)
        stats['Random'].append((budget_max-used,
                                base_avg - calculate_objective_stats(calculate_reduced_heatmap(sol, temp_matrix, tree_species), temp_matrix)[0],
                                base_hot - calculate_objective_stats(calculate_reduced_heatmap(sol, temp_matrix, tree_species), temp_matrix)[1],
                                len(sol),
                                t1-t0))
        # GA
        t0 = time.perf_counter()
        ga_model = run_ga(type_matrix, temp_matrix, budget_max, type_to_num, tree_species)
        t1 = time.perf_counter()
        sol = [(int(ga_model.output_dict['variable'][i]),
                int(ga_model.output_dict['variable'][i+1]),
                int(ga_model.output_dict['variable'][i+2]))
               for i in range(0, len(ga_model.output_dict['variable']), 4)
               if int(ga_model.output_dict['variable'][i+3])==1]
        used = calculate_used_budget(sol, type_matrix, tree_species, type_to_num)
        avg, hot = calculate_objective_stats(calculate_reduced_heatmap(sol, temp_matrix, tree_species), temp_matrix)
        stats['GA'].append((budget_max-used, base_avg-avg, base_hot-hot, len(sol), t1-t0))
        # ACO
        t0 = time.perf_counter()
        sol = run_aco(type_matrix, temp_matrix, type_to_num, budget_max, iterations=50, num_ants=10)
        t1 = time.perf_counter()
        used = calculate_used_budget(sol, type_matrix, tree_species, type_to_num)
        avg, hot = calculate_objective_stats(calculate_reduced_heatmap(sol, temp_matrix, tree_species), temp_matrix)
        stats['ACO'].append((budget_max-used, base_avg-avg, base_hot-hot, len(sol), t1-t0))
        # ACO + CandPh
        t0 = time.perf_counter()
        sol = run_aco(type_matrix, temp_matrix, type_to_num, budget_max, iterations=50, num_ants=10, use_cand_ph=True)
        t1 = time.perf_counter()
        used = calculate_used_budget(sol, type_matrix, tree_species, type_to_num)
        avg, hot = calculate_objective_stats(calculate_reduced_heatmap(sol, temp_matrix, tree_species), temp_matrix)
        stats['ACO + Cand Ph'].append((budget_max-used, base_avg-avg, base_hot-hot, len(sol), t1-t0))
    return stats

# Execute multi-run
runs = 2
stats = evaluate_multiple_runs(runs, type_matrix, temp_matrix, budget_max, type_to_num, tree_species)

# --- 7. Normalize and plot spider chart ---
algs = list(stats.keys())
raw = np.zeros((len(algs), runs, 5))
for i, alg in enumerate(algs):
    for j, entry in enumerate(stats[alg]):
        raw[i, j] = entry    # (budget_rem, delta_avg, delta_hot, num_trees, exec_time)

max_vals = raw.max(axis=(0,1))
mean_vals = raw.mean(axis=1)
norm = mean_vals / max_vals * 100

df_norm = pd.DataFrame(norm, index=algs,
                       columns=['Budget Remaining', 'Delta Avg Temp', 'Delta Hot Temp', 'Num Trees', 'Exec Time'])

# Spider chart
cats = df_norm.columns.tolist()
N = len(cats)
angles = np.linspace(0, 2*np.pi, N, endpoint=False).tolist()
angles += angles[:1]

fig, ax = plt.subplots(figsize=(6,6), subplot_kw=dict(polar=True))
for alg in algs:
    vals = df_norm.loc[alg].tolist()
    vals += vals[:1]
    ax.plot(angles, vals, label=alg)
    ax.fill(angles, vals, alpha=0.25)

ax.set_xticks(angles[:-1])
ax.set_xticklabels(cats, fontsize=10)
ax.set_title("Normalized Spider Chart (0–100)", y=1.08)
ax.legend(loc='upper right', bbox_to_anchor=(1.3,1.1))

plt.tight_layout()
plt.show()
```
